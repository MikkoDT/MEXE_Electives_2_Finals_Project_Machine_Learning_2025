```python
import pandas as pd
dataset = pd.read_csv('games.csv')
dataset.dropna(inplace=True)
```

```python
# Create a new column that shows if the home team won
# If the home team scored more points → put 1, Otherwise → put 0
target_col = 'HOME_TEAM_WINS'

# Create it if it doesn't exist
if target_col not in dataset.columns:
    dataset[target_col] = (dataset['PTS_home'] > dataset['PTS_away']).astype(int)
```

```python
# These are columns that contain numbers (points, rebounds, assists, turnovers, fouls, etc.).We treat them differently from text because they can be used for math.
numerical_cols = [
    'PTS_home', 'FG_PCT_home', 'REB_home', 'AST_home', 'TOV_home', 'PF_home',
    'PTS_away', 'FG_PCT_away', 'REB_away', 'AST_away', 'TOV_away', 'PF_away'
]
```

```python
selected_columns = [
    'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home',
    'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away'
]
```

```python
# These are text-like values, such as team IDs or season labels. They cannot be used directly in machine learning.
categorical_cols = ['HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'SEASON']
```

```python
# If a row has missing numbers (like missing points), delete it. The model can’t learn from incomplete data.
existing_numerical_cols = [col for col in numerical_cols if col in dataset.columns]
dataset = dataset.dropna(subset=existing_numerical_cols + [target_col])
```

```python
# Filter out bad or unrealistic games
dataset = dataset[(dataset['PTS_home'] > 50) & (dataset['PTS_away'] > 50)]
dataset = dataset[(dataset['FG_PCT_home'] > 0) & (dataset['FG_PCT_home'] <= 1)]
dataset = dataset[(dataset['FG_PCT_away'] > 0) & (dataset['FG_PCT_away'] <= 1)]
```

```python
# New columns showing differences between teams: POINT_DIFF, REB_DIFF, AST_DIFF. These help the model understand game performance better.
dataset['POINT_DIFF'] = dataset['PTS_home'] - dataset['PTS_away']
dataset['REB_DIFF']   = dataset['REB_home'] - dataset['REB_away']
dataset['AST_DIFF']   = dataset['AST_home'] - dataset['AST_away']

# Add new features to numerical list
numerical_cols += ['POINT_DIFF','REB_DIFF','AST_DIFF']
```

```python
# Machine learning can't understand text like team IDs. get_dummies turns these into 0/1 columns (one-hot encoding).
dataset_encoded = pd.get_dummies(dataset, columns=categorical_cols, drop_first=True)
```

```python
# import standard scaler

from sklearn.preprocessing import StandardScaler

# Some features have huge numbers (like 1000), some small (like 0.5). Scaling puts everything on a similar range so the model learns better.
scaler = StandardScaler()

# Filter numerical_cols to include only columns present in dataset_encoded
existing_numerical_cols_for_scaling = [col for col in numerical_cols if col in dataset_encoded.columns]
dataset_encoded[existing_numerical_cols_for_scaling] = scaler.fit_transform(dataset_encoded[existing_numerical_cols_for_scaling])
```

```python
# We drop columns that do not help prediction (like game date or game ID).
# Importantly, 'PTS_home', 'PTS_away', and 'POINT_DIFF' are removed to prevent data leakage.
columns_to_drop = ['GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT', 'TEAM_ID_home', 'TEAM_ID_away', 'PTS_home', 'PTS_away', 'POINT_DIFF']

# Ensure only existing columns are dropped
final_columns_to_drop = [col for col in columns_to_drop if col in dataset_encoded.columns]

x = dataset_encoded[selected_columns]
y = dataset_encoded[target_col]
```
