### Importing the dataset

```python
import pandas as pd

dataset = pd.read_csv("cleaned_college_tier_dataset.csv")
dataset.head(10)
```

### Dataset Information

```python
dataset.info()
```

### Getting the inputs and output

```python
X = dataset.drop(columns=['CollegeTier']).values
y = dataset['CollegeTier'].values

X
y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 0
)

X_train
X_test
y_train
y_test
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train
```

### Building the model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 0, max_iter=5000)
```

### Training the model

```python
model.fit(X_train, y_train)
```

### Making the predictions on the test set

```python
y_pred = model.predict(X_test)
y_pred
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

### Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

### Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

###Visualization

###Confusion Matrix Visualization (Heatmap)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

###ROC Curve

```python
from sklearn.metrics import roc_curve, auc

y_test_binary = (y_test == 2).astype(int)
y_pred_prob = model.predict_proba(X_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test_binary, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

- The model achieved high overall accuracy (92.33%).
Performs strongly when predicting Tier 2 colleges due to a large number of examples.

- Logistic Regression still maintains stable decision boundaries even with uneven classes.
